```python
def get_char_set(text_array):
    '''text_array is a pandas Series'''
    str_sets = text_array.apply(lambda x: set(x))
    return sorted(list(set.union(*str_sets)))
```

```python
vocab_set = get_char_set(df['text'])
```

```python
encoder = preprocess.get_mapper(vocab_set)
decoder = preprocess.get_mapper(vocab_set, type='decoder')
```
